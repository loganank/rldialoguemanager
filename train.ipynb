```python
import pandas as pd
from Runner import Runner
```

Grab data from users and turn into dataframe

```python
names = ['message', 'correct_decision']
data = pd.read_csv("user_data.csv", names=names)
data.head()
```

```python
print(len(data))
```

Separate data into two dataframes. One for training (first 5 users) and one for testing (last user). The training data will have 250 samples, and the test data will have 50 samples.

```python
train = data[:250]
test = data[250:]
```

```python
print(len(train))
print(len(test))
```

Train the reinforcement learning model by looping through all of the training samples and providing them to the reinforcement learning agent.

```python
print(train.iloc[0].message)
```

```python
runner = Runner()
for index, row in train.iterrows():
    runner.get_dm_decision(row['message'])
    runner.get_dm_response(row['correct_decision'])
```

Evaluate

```python
total = len(test)
print(total)
```

```python
correct = 0
for index, row in test.iterrows():
    if runner.get_dm_decision(row['message']) == row['correct_decision']:
        correct += 1

print("Accuracy after training: ", correct / total)
```
